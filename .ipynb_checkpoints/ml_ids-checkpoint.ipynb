```python
X, Y = featureSelection.X_new, featureSelection.y
X.loc[:, 'Destination Port']
```

```python
Chi_train = X[featureSelection.chi_feature]
Chi_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score , precision_score , recall_score , f1_score
```

```python
my_model = RandomForestClassifier(n_estimators = 501)
clf_forest = my_model.fit(Chi_train, Y)
#print(clf.get_depth())
```

```python
my_model = MLPClassifier(hidden_layer_sizes = (50, 100, 100, 100, 50) , learning_rate = 'adaptive', verbose = 1)
clf_nn = my_model.fit(Chi_train, Y)
```

```python
clf_nn
```

```python
clf_forest
```

```python
y = clf_nn.predict(Chi_train)
```

```python
print(accuracy_score(Y, y))
print(precision_score(Y, y))
print(recall_score(Y, y))
print(f1_score(Y, y))
```

```python
y1 = clf_forest.predict(Chi_train)
```

```python
print(accuracy_score(Y, y1))
print(precision_score(Y, y1))
print(recall_score(Y, y1))
print(f1_score(Y, y1))
```
